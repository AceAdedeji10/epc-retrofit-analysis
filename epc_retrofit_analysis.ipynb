Importing necessary libraries and reading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('KingsLynnHousingData.xlsx')
df.head()
```

Dropping all irrelevant columns

```python
df = df[[
    'POSTCODE',
    'POSTCODE_DISTRICT',
    'CURRENT_ENERGY_RATING',
    'CURRENT_ENERGY_EFFICIENCY',
    'PROPERTY_TYPE',
    'BUILT_FORM',
    'MAINHEAT_DESCRIPTION'
]].dropna()
```

Cleaning the POSTCODE_DISTRICT column

```python
df['POSTCODE_DISTRICT'] = df['POSTCODE_DISTRICT'].str.strip().str.upper()
```

Converting EPC rating to numeric

```python
rating_map = {
    'A': 1, 'B': 2, 'C': 3,
    'D': 4, 'E': 5, 'F': 6, 'G': 7
}

df['EPC_NUMERIC'] = df['CURRENT_ENERGY_RATING'].map(rating_map)
```

Simulating improvement logic


```python
def simulate_improvement(rating):
    if rating >= 5:   # E, F, G
        return max(1, rating - 2)
    elif rating >= 3: # C, D
        return rating - 1
    else:
        return rating

df['PREDICTED_EPC_NUMERIC'] = df['EPC_NUMERIC'].apply(simulate_improvement)
```

Converting EPC back to letters

```python
reverse_map = {v: k for k, v in rating_map.items()}

df['PREDICTED_EPC_RATING'] = df['PREDICTED_EPC_NUMERIC'].map(reverse_map)
```

Calculating improvement score


```python
df['IMPROVEMENT_SCORE'] = df['EPC_NUMERIC'] - df['PREDICTED_EPC_NUMERIC']
```

Creating Potential Impact Category


```python
def impact_category(x):
    if x >= 2:
        return 'High Impact'
    elif x == 1:
        return 'Medium Impact'
    else:
        return 'Low Impact'

df['IMPACT_CATEGORY'] = df['IMPROVEMENT_SCORE'].apply(impact_category)
```

Checking value counts by Impact Category



```python
df['IMPACT_CATEGORY'].value_counts()
```

Checking Improvement Score by Postcode District

```python
df.groupby('POSTCODE_DISTRICT')['IMPROVEMENT_SCORE'].mean().sort_values(ascending=False)
```

Attaching repair cost to property based on the Improvement score

```python
def estimate_cost(improvement):
    if improvement >= 2:
        return 7000
    elif improvement == 1:
        return 3000
    else:
        return 0

df['ESTIMATED_COST'] = df['IMPROVEMENT_SCORE'].apply(estimate_cost)
```

Defining the total budget amount

```python
BUDGET = 5_000_000  # £5 million
```

Priortizing homes based on improvement potential and current EPC rating


```python
df_sorted = df.sort_values(
    by=['IMPROVEMENT_SCORE', 'EPC_NUMERIC'],
    ascending=[False, False]
)
```

Simulating spending budget

```python
total_spent = 0
selected_rows = []

for _, row in df_sorted.iterrows():
    cost = row['ESTIMATED_COST']

    if total_spent + cost <= BUDGET and cost > 0:
        selected_rows.append(row)
        total_spent += cost

selected_df = pd.DataFrame(selected_rows)
```

To measure the impact


```python
homes_upgraded = len(selected_df)

avg_improvement = selected_df['IMPROVEMENT_SCORE'].mean()

total_improvement = selected_df['IMPROVEMENT_SCORE'].sum()
```

Calculating EPC averages before and after

```python
before_avg = selected_df['EPC_NUMERIC'].mean()
after_avg = selected_df['PREDICTED_EPC_NUMERIC'].mean()
```

Printing results

```python
print("Total Budget:", BUDGET)
print("Total Spent:", total_spent)
print("Homes Upgraded:", homes_upgraded)
print("Average Improvement:", avg_improvement)
```
